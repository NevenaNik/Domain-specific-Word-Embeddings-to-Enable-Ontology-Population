### Read in csv

```python
pathTrain = "/home/hiwi/Dokumente/masterthesis/data/trained/"
pathEval = "/home/hiwi/Dokumente/masterthesis/data/eval/"
dataTrain = pd.read_csv(pathTrain+"word2vec_trainingLog.csv", sep=";")
dataEval = pd.read_csv(pathEval+"results_part1_01b_12_12.csv", sep=";")
data = pd.concat([dataTrain, dataEval], axis=1, sort=False)
data = data.loc[:,~data.columns.duplicated()]
```

```python
data["id"] = data["id"].astype(str)
data["reduction"] = data["reduction"].astype(str)
data["rare"] = data["rare"].astype(str)
data["shuffle"] = data["shuffle"].astype(str)
```

```python
data = data.replace({'groupRare': {1: "shuffle:False, ngram:Man", 2: "shuffle:False, ngram:Train", 3:"shuffle:True, ngram:Man", 4:"shuffle:True, ngram:Train"}})
data = data.replace({'groupShuffle': {1: "rare:False, ngram:Man", 2: "rare:False, ngram:Train", 3:"rare:True, ngram:Man", 4:"rare:True, ngram:Train"}})
data = data.replace({'groupNgrams': {1: "rare:False, shuffle:False", 2: "rare:False, shuffle:True", 3:"rare:True, shuffle:False", 4:"rare:True, shuffle:True"}})
```

```python
data
```

```python
data["groupRare"] = data.groupby(["shuffle", "ngrams"]).ngroup()
data["groupRare"] = data["groupRare"] + 1
data["groupShuffle"] = data.groupby(["rare", "ngrams"]).ngroup()
data["groupShuffle"] = data["groupShuffle"] +1
data["groupShuffle"] = data["groupShuffle"].astype(str)
data["groupNgrams"] = data.groupby(["rare", "shuffle"]).ngroup()
data["groupNgrams"] = data["groupNgrams"] +1
```

```python
# Shuffle vs. No-Shuffle

methods = ["True", "False"]
markers = ["hex", "triangle"]

p = figure(title="Training Part 1: Evaluation results")
p.xaxis.axis_label = "Constant Groups"
p.yaxis.axis_label = "Eval score"

p.scatter("groupShuffle", "evalScore", source=data, legend_field="shuffle", fill_alpha=0.4, size=10,
            marker=factor_mark("shuffle", markers, methods),
            color=factor_cmap("shuffle", "Category10_3", methods))

p.add_layout(p.legend[0], "right")

show(p)
```
